## Reading the data into Python
Using the pd.read_csv() function. 
College is now a dataframe containing our data.

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt


college = pd.read_csv('College.csv')
```

## Inspect and set index

First we just check out the original dataframe. We see that the first column is called 'Unnamed:0' and containg the colleges/university.
With college2 we put the first column as index.
With college3 we instead rename the first column to 'College', and also set this to be the index. 
We then overriden our original dataframe with the contents of college3.

```python
college 

college2 = pd.read_csv('College.csv', index_col=0) 

college3 = college.rename({'Unnamed: 0': 'College'}, axis=1) 
college3 = college3.set_index('College') 
college = college3 
college 
```

## Scatterplot matrix for selected columns 

Here we use pd.plotting.scatter_matrix() and pass in the columns 'Top10perc', 'Apps', 'Enroll' in order to examine them.  A scatterplot matrix shows the relationships between multiple numeric variables in a dataset. On the diagonal we see each variable in a histogram each. The rest are scatterplots with 2 of the variables each.  

What do we discover?
There seems to be a correlation between Applications and Enrollment - we can detect a pattern in how the points are arranged. 


```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

## Boxplot of variables 'Private' vs 'Outstate'  
It seems that at private colleges there are more students from out of state, while at public colleges there are less. The median of outstate students differs with around 10 000 students between public and private colleges. It could be that more students travel for studies when it comes to the more prestiqgous, private colleges compared to the public colleges more frequented by locals.

```python
college.boxplot(column='Outstate', by='Private', figsize=(6,4)) 
```

## Creating the column 'Elite'
by using pd.cut() function.

'Elite' (yes/no) weather the proportion of students that were in the top 10% is more or less than 50%

Here we see that at elite(no) colleges there are lot of outstate students. So presumably students travel out of state for college even for non-elite colleges, more so than for public colleges. Though many students more travel out of state for college when it is an elite rather than non-elite college.
Also the median has a large gap here.
Another noticible aspect is that for elite colleges the lower quartile is at above 5000, meaning that the lowest amount of out of state students is a lot bigger than at non-elite.


```python
college ['Elite'] = pd.cut(college['Top10perc'], [-1,50,100], labels =['No', 'Yes'])
college['Elite'].value_counts()
college['Elite']

college.boxplot(column='Outstate', by='Elite', figsize=(6,4))  
```

## 4 histogram subplots of quantitative variables 
We see that it is most common to spend around 500 on books.
